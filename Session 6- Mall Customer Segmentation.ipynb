# Mall Customer Segmentation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer = pd.read_csv("Mall_Customers.csv")
customer
```

## Preprocessing

```python
# Dropping CustomerID

customer = customer.drop(["CustomerID"], axis = 1)
customer
```

```python
# Converting Categorical to Numeric

cust = pd.get_dummies(customer)
cust
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(cust)
X
```

```python
# Converting to DF

X = pd.DataFrame(X, columns = cust.columns)
X
```

## Dimensionality Reduction using TSNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components = 2, random_state = 100)
```

```python
type(tsne)
```

```python
X_tsne = tsne.fit_transform(X)
X_tsne
```

```python
# Visualization

plt.scatter(X_tsne[:, 0], X_tsne[:, 1]);
```

### Findings

1. Cluster formation is clear
2. Number of clusters may be 2

## Clustering usinf GMM

```python
from sklearn.mixture import GaussianMixture
n_comps = np.arange(1, 20, 1)
aic_score = []
bic_score = []
for n in n_comps:
    model = GaussianMixture(n_components = n, random_state = 100, n_init = 5)
    model.fit(X)
    aic_score.append(model.aic(X))
    bic_score.append(model.bic(X))
```

```python
print(aic_score)
print(bic_score)
```

```python
plt.plot(n_comps, aic_score, c = "r", label = "AIC")
plt.plot(n_comps, bic_score, c = "b", label = "bIC")
plt.legend();
```

## Remarks

Number of clusters is 2

### Creating a GM model with 2 clusters

```python
gm = GaussianMixture(n_components = 2, random_state = 100, n_init = 5)
```

```python
gm.fit(X)
```

```python
# prediction

pred = gm.predict(X)
pred
```

```python
gm.weights_
```

## Adding cluster labels to DF

```python
customer["Label"] = pred
customer
```
